```python
import cv2
import matplotlib.pyplot as plt
import pytesseract
import easyocr
import numpy as np
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
img = cv2.imread(r"F:\ai\ASDC\OCR\dataset2\dataset2\WhatsApp Image 2023-10-26 at 3.38.21 PM (2).jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
i = img[250:370, 1300:1560]
plt.imshow(img[250:370, 1300:1560])
print(np.max(i), np.min(i), np.average(i), sep=' ==== ')
```

```python
first_gray = cv2.cvtColor(i, cv2.COLOR_RGB2GRAY)
image_array = np.array(first_gray)

# Create a mask where pixel values are 50 or less
mask = image_array <= 96

# Apply the mask to the image array
masked_image = np.zeros_like(image_array)
masked_image[mask] = image_array[mask]
l = cv2.threshold(masked_image, 0, 255, cv2.THRESH_BINARY_INV)[1]

plt.imshow(l, cmap='gray')
```

```python
first_gray = cv2.cvtColor(img[490:550, 350:850], cv2.COLOR_BGR2GRAY)
(first_thresh, first_blackAndWhite) = cv2.threshold(first_gray, 80, 255, cv2.THRESH_BINARY)

blur = cv2.GaussianBlur(first_blackAndWhite,(5,5),10)
ret3,th_first = cv2.threshold(blur, 30, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# Morphological operations (optional, helps in cleaning up the image)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
morphology = cv2.morphologyEx(th_first, cv2.MORPH_CLOSE, kernel, iterations=1)
plt.imshow(morphology, cmap='gray')
```
